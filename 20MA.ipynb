```python
import numpy as np
import yfinance as yf
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import datetime as dt
from datetime import date
from binance import Client
```

```python

ma_1 = 20
ma_2 = 100
ticker = 'TSLA'
```

```python

start = dt.datetime.now() - dt.timedelta(days=365*3)
end = dt.datetime.now()
```

```python
data= yf.download(ticker,start,end,interval='1d')
data.head()
```

```python
plt.style.use('dark_background')
plt.rcParams['figure.figsize']=(40,20)
plt.rcParams.update({'font.size':40})

data [f'SMA_{ma_1}'] =data['Adj Close'].rolling(window=ma_1).mean()
data [f'SMA_{ma_2}'] =data['Adj Close'].rolling(window=ma_2).mean()

plt.plot(data['Adj Close'], label=ticker, color='lightgray',linewidth=3)
plt.plot(data[f'SMA_{ma_1}'],label=f'SMA_{ma_1}', color='orange',linewidth=3)
plt.plot(data[f'SMA_{ma_2}'],label=f'SMA_{ma_2}', color='purple',linewidth=3)


plt.legend(loc='upper left')
plt.show()
```

```python

Dif = (data [f'SMA_{ma_1}'] - data['Adj Close'])/data['Adj Close']

plt.plot(Dif, label="% Difference", color='lightgray',linewidth=3)



plt.legend(loc='upper left')
plt.show()
```
